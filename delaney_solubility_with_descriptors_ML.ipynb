**ML Project**




*Load deta*

```python
import pandas as pd
df = pd.read_csv( 'delaney_solubility_with_descriptors.csv' )
df
```

**|Data prepration**

*Data Seperation x and y*

```python
y = df['logS']
y
```

```python
x = df.drop('logS', axis=1)
x
```

*Data splitting(Testing and Traning)*

```python
import sklearn
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=100)
```

```python
x_train
```

```python
x_test
```


**Model Building**



*Linear regression*

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train, y_train)
```

Applying the model to make pridiction

```python
y_lr_train_pred = lr.predict(x_train)
y_lr_test_pred = lr.predict(x_test)
```

```python
print(y_lr_train_pred, y_lr_test_pred)
```

**Random Forest**

*Traning the model*

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(max_depth=2, random_state=100)
rf.fit(x_train, y_train)
```

*applying the model to make the prediction*

```python
y_rf_train_pred = rf.predict(x_train)
y_rf_test_pred = rf.predict(x_test)
```

*Evalution model performence*

```python
import sklearn
from sklearn.metrics import mean_squared_error, r2_score

rf_train_mse = mean_squared_error(y_train, y_rf_train_pred)
rf_train_r2 = r2_score(y_train, y_rf_train_pred)

rf_test_mse = mean_squared_error(y_test, y_rf_test_pred)
rf_test_r2 = r2_score(y_test, y_rf_test_pred)
```

```python
rf_results = pd.DataFrame(['Liner Regression', rf_train_mse, rf_train_r2, rf_test_mse, rf_test_r2]).transpose()
rf_results.columns = ['Method', 'Traning MSE', 'Traning R2', 'Test MSE', 'Test R2']
```

**Data visualization of prediction result**

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(5, 5))

# Scatter plot with alpha for transparency
plt.scatter(x=y_train, y=y_lr_train_pred, c="#006400", alpha=0.3)

# Perform linear regression (replace with your actual model training)
z = np.polyfit(y_train, y_lr_train_pred, 1)  # Assuming y_train and y_lr_train_pred have the same length

# Evaluate the fitted polynomial for the training data
p = np.polyval(z, y_train)  # Use np.polyval for polynomial evaluation

# Plot the fitted line in red
plt.plot(y_train, p, color='#000000', label='Fitted Line')  # Add label for clarity

plt.ylabel('Predicted LogS')
plt.xlabel('Experimental LogS')
plt.legend()  # Show the legend for the fitted line

plt.show()
```
